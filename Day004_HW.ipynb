### 比較一下範例檔案中的「r.text」與「json.loads(r.text)」讀出來的內容有什麼差異

```python
import requests

r = requests.get('https://cat-fact.herokuapp.com/facts')
r.text
#r.text是把屬性和資料內容都以字串的型態顯示出來("")

import json
json.loads(r.text)
#json.loads(r.text)會自動把一層一層的資料擺好呈現
```
